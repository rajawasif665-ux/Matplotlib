```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Countries
countries = [
    "USA", "China", "India", "Russia", "Japan", "Germany", "Brazil", "UK", "France", "Canada",
    "Australia", "South_Africa", "Mexico", "Indonesia", "Saudi_Arabia", "Argentina", "Nigeria",
    "Egypt", "Turkey", "Iran"
]  
# Years 1800 to 2014
years = np.arange(1800, 2014) 

# Temperature rise for each country
data = {}
for country in countries:
    temps = -0.5 + (years - 1800)/213 * np.random.uniform(1, 2) + np.random.normal(0, 0.1, len(years))
    data[country] = temps

df = pd.DataFrame(data, index=years)

# Total temperature rise per country
temp_rise = df.loc[2013] - df.loc[1800]

# 2. Plotting
plt.figure(figsize=(18, 12)) # WIDTH 18 AND LENGTH 12

# Line + Scatter Plot 
plt.subplot(2,2,1)
for country in countries:
    plt.plot(df.index, df[country], label=country)
    plt.scatter(df.index[::30], df[country].values[::30], s=10)  # scatter every 30 years
plt.title("Temperature Anomalies (1800-2013)")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.grid(True)
plt.legend(fontsize=8, ncol=4)

# Bar Chart 
plt.subplot(2,2,2)
plt.bar(countries, temp_rise, color='skyblue')
plt.title("Total Temperature Rise per Country")
plt.ylabel(" Temperature (C)")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Pie Chart 
plt.subplot(2,2,3)
# Only positive value for pie chart
shifted = temp_rise - temp_rise.min() + 0.01
plt.pie(shifted, labels=countries, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 8})
plt.title("Share of Temperature Rise (shifted)")

plt.tight_layout()
plt.show()
```
